**1. Скільки чоловіків и жінок (ознака *sex*) представлено в цьому наборі даних?**

````python
# dataframe version:
# gender_counts = data["sex"].value_counts()

female_amount = len(data[data["sex"] == "Female"])
male_amount = len(data[data["sex"] == "Male"])
print(f"Female amount: {female_amount}")
print(f"Male amount: {male_amount}")
````

**2. Який середній вік (ознака *age*) жінок?**

````python
average_age_female = data[data["sex"] == "Female"]["age"].mean()
print(f'Average female age: {average_age_female}')
````

**3. Яка доля громадян Німеччини (ознака *native-country*)?**

````python
germany_native_count = len(data[data["native-country"] == "Germany"])
germany_native_ratio = germany_native_count / len(data)
print(f'Germany native ratio: {germany_native_ratio}')
````

**4-5. Які средні значення та средньоквадратичні відхилення віку тих, хто отримує більше 50K на рік (ознака *salary*) і тих, хто отримує менше 50K на рік?**

````python
# dataframe version:
# age_stats = data.groupby("salary")["age"].agg(["mean", "std"])
high_income = data[data["salary"] == ">50K"]
low_income = data[data["salary"] == "<=50K"]

mean_high = high_income["age"].mean()
std_high = high_income["age"].std()

mean_low = low_income["age"].mean()
std_low = low_income["age"].std()
print(f"Those earning >50K: average age = {mean_high}, standard deviation = {std_high}")
print(f"Those earning <=50K: average age = {mean_low}, standard deviation = {std_low}")
````

**6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту? (ознака *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

````python
higher_education = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]
has_higher_education = high_income["education"].isin(higher_education)

if has_higher_education.all():
    print("True, all people earning >50K have at least higher education.")
else:
    print("False, not all people earning >50K have higher education.")
````

**7. Порахуйте статистику віку для кожної раси (ознака *race*) і кожної статі. Використайте *groupby* и *describe*. Знайдіть таким сином максимальний вік чоловіків раси *Amer-Indian-Eskimo*.**

````python
age_stats = data.groupby(["race", "sex"])["age"].describe()

max_age_amer_indian_eskimo_male = data[(data["race"] == "Amer-Indian-Eskimo") & (data["sex"] == "Male")]["age"].max()
print(f'Max age of Amer-Indian-Eskimo males: {max_age_amer_indian_eskimo_male}')
````

**8. Серед кого більша доля багатіших (>50K): серед одружених чи не одружених чоловіків (ознака *marital-status*)? За одружених вважаємо тих, у кого *marital-status* починається із *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), всі інші - не одружені.**

````python
men = data[data["sex"] == "Male"]

married_statuses = ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
is_married = men["marital-status"].isin(married_statuses)

married_men_rich_ratio = (men[is_married]["salary"].eq(">50K")).shape[0] / men.shape[0]
unmarried_men_rich_ratio = (men[~is_married]["salary"].eq(">50K")).shape[0] / men.shape[0]

if married_men_rich_ratio > unmarried_men_rich_ratio:
  print(f'Married rich men ratio {married_rich_ratio} is more than unmarried rich men ratio {unmarried_rich_ratio}')
else:
  print(f'Unmarried rich men ratio {unmarried_rich_ratio} is more than married rich men ratio {married_rich_ratio}')
````

**9. Яку максимальну кількість годин в тиждень працює людина (ознака *hours-per-week*)? Скільки людей працюють таку кількість годин і який відсоток серед них багатих?**

````python
max_hours = data["hours-per-week"].max()
num_people_max_hours = data[data["hours-per-week"] == max_hours].shape[0]
rich_ratio = data[(data["hours-per-week"] == max_hours) & (data["salary"] == ">50K")].shape[0] / num_people_max_hours

print(f"Maximum number of hours worked per week: {max_hours}")
print(f"Number of people working maximum hours: {num_people_max_hours}")
print(f"Percentage of reach people: {rich_ratio:.2%}")
````
