```python
# 1. Знайти в датасеті таргет та видалити цю колонку з датасету (видаляти за індексом)
a =np.genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', dtype=float, delimiter=',', usecols=(0,1,2,3))
```

```python
# 2. Перетворити колонки, що залишились в 2D масив (або впевнитись, що це уже 2D масив)
a.shape
```

```python
# 3. Порахувати mean, median, standard deviation для 1-ї колонки
column_1_mean = np.mean(a[:, 0])
print(f'mean: {column_1_mean}')
column_1_median = np.median(a.T[0])
print(f'median: {column_1_median}')
deviation_1_column = np.std(a, axis=0)[0]
print(f'deviation: {deviation_1_column}')
```

```python
# 4. Вставити 20 значень np.nan на випадкові позиції в масиві
#   (при використанні звичайного рандому можуть накластись позиції, тому знайти рішення, яке гарантує 20 унікальних позицій)
a.flat[np.random.choice(a.size, 20, replace=False)] = np.nan
a
```

```python
# 5. Знайти позиції вставлених значень np.nan в 1-й колонці
print(f'Nan positions in the first column: \n{np.where(np.isnan(a[:, 0]))}')
```

```python
# 6. Відфільтрувати массив за умовою: значення в 3-й колонці > 1.5 та значения в 1-й колонці < 5.0 (зберегти у іншу змінну)
filtered_array = a[(a[:, 2] > 1.5) & (a[:, 0] < 5.0)]
filtered_array
```

```python
# 7. Замінити всі значення np.nan на 0
a = np.nan_to_num(a, nan=0)
```

```python
# 8. Порахувати всі унікальні значення в массиві та вивести їх разом із кількістю
unique_values, counts = np.unique(a, return_counts=True)
print(f"Unique values and count:\n{np.column_stack((unique_values, counts))}")
```

```python
# 9. Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)
mid = a.shape[0] // 2
top_array_part = a[:mid, :]
bottom_array_part = a[mid:, :]
print(f'Top array part:\n{top_array_part}')
print(f'Botttom array part:\n{bottom_array_part}')
```

```python
# 10. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням
sorted_top_array_part_asc = top_array_part[top_array_part[:, 0].argsort()]
sorted_bottom_array_part_desc = bottom_array_part[bottom_array_part[:, 0].argsort()[::-1]]
print(f"Sorted top by asc: \n{sorted_top_array_part_asc}")
print(f"Sorted top by desc: \n{sorted_bottom_array_part_desc}")
```

```python
# 11. Зібрати обидва массиви в одне ціле
np.vstack((top_array_part, bottom_array_part))
```

```python
# 12. Знайти найбільш часто повторюване значення в массиві
unique, indices = np.unique(a.flatten(), return_inverse=True)
most_frequent = unique[np.argmax(np.bincount(indices))]
most_frequent
```

```python
# 13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.
def modify_column(arr, col_index):
    mean_value = np.mean(arr[:, col_index])

    for i in range(arr.shape[0]):
        if arr[i, col_index] < mean_value:
            arr[i, col_index] *= 2
        else:
            arr[i, col_index] /= 4

    return arr
```

```python
# 14. Застосувати отриману функцію до 3-ї колонки
modified_array = modify_column(a, 3)
modified_array
```
